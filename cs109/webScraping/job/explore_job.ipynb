```python
from datetime import datetime, timedelta 
import pandas as pd
import glob, os
import numpy as np

def ex_filter(title, exList):
    for keyword in exList:
        if(keyword.lower() in title.lower()):
            #print("------------>" + keyword + " found in " + title)
            return 0
    return 1

os.chdir("/media/zhou/DATA/zhousu/git/github_new/DataLab/cs109/webScraping/job/data")
dfList = []
for file in glob.glob("*.csv"):
    mtime = os.path.getmtime(file)
    #print(mtime)
    now = datetime.now()
    before = now + timedelta(hours = -9)
    ts = before.timestamp()
    #print(ts)    
    if mtime > ts:
        print(file)
        df1 = pd.read_csv(file)
        dfList.append(df1)

df = pd.concat(dfList)

pd.set_option('display.max_colwidth', -1)

excludeCompany = "Valtech|Peytz|UNOPS|Mercell|GlobeSearch|Laerdal|Commentor|Game|Academy|Handiscover"
excludeCompany += "|KMD|Bloom|Maersk|BLAST|DevFactory|Proper|Netcom|DTU|Portchain|Enghouse|Kind|Neo4j"

excludeTitle = "Scrum|Change|WordPress|SRE|test|Kotlin|iOS|Admin|30_000|Junior"
excludeLocation="Lund"
excludeLink="460585|2326315982|457937"

df = df[ 
    #(~df['Location'].str.contains("Malm", case=False)) 
    (~df['Company'].str.contains(excludeCompany, case=False)) 
   & (~df['Title'].str.contains(excludeTitle, case=False)) 
   & (~df['Location'].str.contains(excludeLocation, case=False))
   & (~df['Link'].str.contains(excludeLink, case=False))
  ]

df.drop_duplicates(subset =['Company', 'Title'], keep = 'last', inplace = True) 

df.sort_values(by=['Location', 'Company', 'Title'])
```
